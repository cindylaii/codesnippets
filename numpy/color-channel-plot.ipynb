```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

### Option 1

```python
img = cv2.imread('images/my_image2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
zeros = np.zeros((img.shape[0], img.shape[1], 3), dtype='uint8')
```

```python
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]
```

```python
img_red = np.copy(zeros)
img_red[:,:,0] = red

img_green = np.copy(zeros)
img_green[:,:,1] = green

img_blue = np.copy(zeros)
img_blue[:,:,2] = blue
```

```python
f, axes = plt.subplots(1,4, figsize=(20,10))
axes[0].imshow(img)
axes[1].imshow(img_red)
axes[2].imshow(img_green)
axes[3].imshow(img_blue)
```

### Option 2

```python
img = cv2.imread('images/my_image2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def red_channel_image(img):
    red = img[:,:,0]
    zero = np.zeros((red.shape[0], red.shape[1], 2), dtype='uint8')
    img_red = np.dstack((red, zero))
    return img_red

def green_channel_image(img):
    green = img[:,:,1]
    zero = np.zeros((green.shape[0], green.shape[1], 1), dtype='uint8')
    tmp = np.dstack((zero, green))
    img_green = np.dstack((tmp, zero))
    return img_green

def blue_channel_image(img):
    blue = img[:,:,2]
    zero = np.zeros((blue.shape[0], blue.shape[1], 2), dtype='uint8')
    img_blue = np.dstack((zero,blue))
    return img_blue
```

```python
f, axes = plt.subplots(1,4, figsize=(20,10))
axes[0].imshow(img)
axes[1].imshow(red_channel_image(img))
axes[2].imshow(green_channel_image(img))
axes[3].imshow(blue_channel_image(img))
```

### Option 3

```python
img = cv2.imread('images/my_image2.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def red_channel_image(img):
    red = img[:,:,0]
    zeros = np.zeros((red.shape[0], red.shape[1], 1), dtype='uint8')
    tmp = np.append(np.atleast_3d(red), zeros, axis=2)
    img_red = np.append(tmp, zeros, axis=2)
    return img_red

def green_channel_image(img):
    green = img[:,:,1]
    zeros = np.zeros((green.shape[0], green.shape[1], 1), dtype='uint8')
    tmp = np.append(zeros, np.atleast_3d(green), axis=2)
    img_green = np.append(tmp, zeros, axis=2)
    return img_green

def blue_channel_image(img):
    blue = img[:,:,2]
    zeros = np.zeros((blue.shape[0], blue.shape[1], 2), dtype='uint8')
    img_blue = np.append(zeros, np.atleast_3d(blue), axis=2)
    return img_blue
```

```python
f, axes = plt.subplots(1,4, figsize=(20,10))

axes[0].set_title("original")
axes[0].imshow(img)

axes[1].set_title("red")
axes[1].imshow(red_channel_image(img))

axes[2].set_title("green")
axes[2].imshow(green_channel_image(img))

axes[3].set_title("blue")
axes[3].imshow(blue_channel_image(img))
```
